```python
def proc_prod(prod_path='../data/producto_tabla.csv'):
    '''
    '''
    df_prod = pd.read_csv(prod_path)
    df_prod['brand'] = df_prod['NombreProducto'].str.split(' ').str[-2]
    
    weight =  df_prod['NombreProducto'].str.extract('(\d+)(g|Kg|ml)',expand=False)
    weight[0] = weight[0].astype('float')
    df_prod['weight'] =  np.where(weight[1].isin(['Kg']), 1000 * weight[0],weight[0])
    
    
    df_prod['other'] =  df_prod['NombreProducto'].str.extract('Va(i)?nill',expand=False)
    df_prod.to_csv('../data/preprocessed_products.csv', index = False)
    #return df_prod
proc_prod()
```
